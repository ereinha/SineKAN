```python
from sine_kan import *

import numpy as np
import torch
import torch.optim as optim
import matplotlib.pyplot as plt
```

```python
device = torch.device('cuda:0')

def tanh_plus_well(x):
    # f(x) = \tanh(10x + 0.5 + \text{ReLU}(x^2) \cdot 10)
    return torch.tanh(10*x+0.5+torch.nn.functional.relu(x**2)*10)

def sin_cos_exp(x):
    # g(x) = \sin(x) + \cos(5x) \cdot e^{-x^2} + \text{ReLU}(x - 0.5)
    return torch.sin(x) + torch.cos(5*x) * torch.exp(-x**2) + torch.nn.functional.relu(x-0.5)

def sigmoid_plus_relu(x):
    # h(x) = \sigma(3x) + \text{ReLU}(\sin(2x) + x^3)
    # where \sigma(x) = \frac{1}{1 + e^{-x}} is the sigmoid function.
    return torch.sigmoid(3*x) + torch.nn.functional.relu(torch.sin(2*x) + x**3)

def tanh_relu_cos(x):
    # k(x) = \tanh(5x - 2) + 3 \cdot \text{ReLU}(\cos(x^2))
    return torch.tanh(5*x - 2) + torch.nn.functional.relu(torch.cos(x**2)) * 3

def softplus_square_tanh(x):
    # m(x) = \text{Softplus}(x^2 - 1) + \tanh(4x + 0.1)
    # where \text{Softplus}(x) = \log(1 + e^x)
    return torch.nn.functional.softplus(x**2 - 1) + torch.tanh(4*x + 0.1)

def exponential_relu_tanh(x):
    # n(x) = e^{-x^2 + 0.3x} + \text{ReLU}(\tanh(2x - 1))
    return torch.exp(-x**2 + 0.3*x) + torch.nn.functional.relu(torch.tanh(2*x - 1))

functions = [tanh_plus_well, sin_cos_exp, sigmoid_plus_relu, tanh_relu_cos, softplus_square_tanh, exponential_relu_tanh]

latex_strings = ["f(x) = \tanh(10x + 0.5 + \text{ReLU}(x^2) \cdot 10)",
                 "g(x) = \sin(x) + \cos(5x) \cdot e^{-x^2} + \text{ReLU}(x - 0.5)",
                 "h(x) = \sigma(3x) + \text{ReLU}(\sin(2x) + x^3)",
                 "k(x) = \tanh(5x - 2) + 3 \cdot \text{ReLU}(\cos(x^2))",
                 "m(x) = \text{Softplus}(x^2 - 1) + \tanh(4x + 0.1)",
                 "n(x) = e^{-x^2 + 0.3x} + \text{ReLU}(\tanh(2x - 1))"
                 ]

all_y_vals = []

epochs, batch = 1000, 100
for function in functions:
    torch.manual_seed(42)
    op = SineKANLayer(1, 1, grid_size=100, norm_freq=False).to(device)

    # %%
    opt = torch.optim.Adam(op.parameters(), lr=0.01)
    criterion = torch.nn.MSELoss()
    for ep in range(epochs):
        x = (torch.rand(batch).to(device)*2 - 1).unsqueeze(-1)
        y0 = function(x).to(device)

        opt.zero_grad()
        y = op(x)
        loss = criterion(y, y0)
        loss.backward()
        opt.step()
        print(f"Epoch: {ep}, MSE_loss: {loss.item()}")


    # %%
    import matplotlib.pyplot as plt
    # %%
    with torch.no_grad():
        x = torch.linspace(-1., 1., 200).to(device).unsqueeze(-1)
        y0 = function(x).to(device)
        y = op(x)

    all_y_vals.append(y)
    plt.plot(x.cpu(), y0.cpu())
    plt.plot(x.cpu(), y.cpu())
    plt.show()
```

```python
latex_strings = ["f(x) = tanh(10x + 0.5 + ReLU(x^2) \cdot 10)",
                 "g(x) = \sin(x) + \cos(5x) \cdot e^{-x^2} + ReLU(x - 0.5)",
                 "h(x) = \sigma(3x) + ReLU(\sin(2x) + x^3)",
                 "k(x) = tanh(5x - 2) + 3 \cdot ReLU(\cos(x^2))",
                 "m(x) = Softplus(x^2 - 1) + tanh(4x + 0.1)",
                 "n(x) = e^{-x^2 + 0.3x} + ReLU(tanh(2x - 1))"
                 ]

plt.rcParams.update({'font.size': 14})
fig, axs = plt.subplots(3, 2, figsize=(12, 8), dpi=600)
x = np.linspace(-1., 1., 200)
for i, (y_vals, latex_str, function) in enumerate(zip(all_y_vals, latex_strings, functions)):
    ax = axs[i // 2, i % 2]
    ax.plot(x, y_vals.to('cpu').numpy(), label='SineKAN Fit', linewidth=3)
    ax.plot(x, function(torch.from_numpy(x)).to('cpu').numpy(), label=f"${latex_str}$", linestyle='--', linewidth=3)
    ax.text(0.05, 0.95, f"({chr(97+i)})", transform=ax.transAxes, fontsize=28,
            verticalalignment='top')
    ax.set_xlim(-1, 1)
    ax.legend(framealpha=0.3, loc='lower right')
    ax.grid(True)

plt.tight_layout()
plt.savefig('./saved_results/universal_approx.jpg')
```
